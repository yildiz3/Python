# adjustText Demo

_"A small library for automatic adjustment of text position in matplotlib plots to minimize overlaps."_

<br>

Check out the adjustText [GitHub page](https://github.com/Phlya/adjustText).

Created by [Ilya Flyamer](https://github.com/Phlya).

```python
#!pip install adjustText
```

```python
from adjustText import adjust_text

import numpy as np
from matplotlib import pyplot as plt
```

## Basic Example

```python
np.random.seed(33)

x, y = np.random.random((2, 20))
```

Without `adjustText` annotations often run into each other or the points they represent.

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]
```

Using `adjustText` prevents text overlaps and text that falls outside of plotting area.

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts)
```

## Additional Options

### Arrows

```python
np.random.seed(33)

x, y = np.random.random((2, 20))
```

Sometimes the labels end up drifting a bit away from the points.

See Point 18 and Point 1 below.

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts)
```

If necessary, you can add arrows with the keyword argument `arrowprops`.

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts, arrowprops={'arrowstyle': 'fancy', 'color': 'crimson'})
```

### Avoiding Large Markers

If you'd like those labels to be even further from the points, use the `expand_points` option.

```python
plt.scatter(x, y, s=100)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts, expand_points=(1.3, 1.3))
```

### Saving Iterations

`adjustText` prints out the number of iterations completed, but you can also save the iteration figures themselves.

```python
#!mkdir scatter_iter
```

```python
plt.scatter(x, y)

texts = [plt.text(x[i], y[i], f'Point {i}') for i in range(len(x))]

adjust_text(texts, save_steps=True, save_prefix='scatter_iter/')
```
