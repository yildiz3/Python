```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# 1. Patterns

```python
values = [10, 20, 30]
```

```python
plt.figure(figsize=(4, 4))
plt.bar(range(len(values)), values, facecolor='lightblue')
plt.xticks([0, 1, 2])
sns.despine()

#plt.savefig('patterns_nopat.svg');
```

```python
plt.rcParams['hatch.linewidth'] = 3
```

```python
plt.figure(figsize=(4, 4))
plt.bar(range(len(values)), values, facecolor='lightblue', hatch='/')
plt.xticks([0, 1, 2])
sns.despine()

#plt.savefig('patterns_withpat.svg');
```

# 2. Shapes

```python
np.random.seed(1204145)
X = np.random.normal(5, 2, size=30)
Y = X + np.random.normal(2, 2, size=30)
```

```python
from matplotlib.patches import Rectangle
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X, Y, s=100, alpha=0.6, edgecolor='black')
plt.xlim(0, 8.5)
plt.ylim(0, 12.5)
sns.despine()

#plt.savefig('shapes_norect.svg');
```

```python
rect = Rectangle((3.5, 5.8), 4.75, 6.2, lw=4, color='red', fill=False)

plt.figure(figsize=(4,4))
plt.scatter(X, Y, s=100, alpha=0.6, edgecolor='black')
plt.gca().add_patch(rect)
plt.xlim(0, 8.5)
plt.ylim(0, 12.5)
sns.despine()

#plt.savefig('shapes_rect.svg');
```

# 3. Baseline

```python
grades = [85, 68, 94]
```

```python
plt.figure(figsize=(4, 4))
plt.bar(range(len(grades)), grades, facecolor='mistyrose', edgecolor='gray')
plt.xticks(range(len(grades)), ['Aly', 'Ben', 'Cole'])
plt.ylim(0, 110)
sns.despine()

#plt.savefig('baseline_noline.svg');
```

# 4. Shading

```python
from datetime import datetime
```

```python
dates = [datetime(2019, 11, 15), datetime(2019, 12, 15), datetime(2020, 1, 15),
         datetime(2020, 2, 15), datetime(2020, 3, 15)]
prices = [110.40, 125.22, 78.30, 66.90, 78.05]
```

```python
plt.figure(figsize=(4, 4))
pd.Series(prices, index=dates).plot(marker='o', markersize=10)
plt.xticks([datetime(2019, 12, 1), datetime(2020, 1, 1), 
            datetime(2020, 2, 1),datetime(2020, 3, 1)], 
           ['Dec\n2019', 'Jan\n2020', 'Feb\n2020', 'March\n2020'], rotation=0, ha='center') 
plt.xlim(datetime(2019, 11, 10), datetime(2020, 3, 20))
plt.ylim(50, 150)

sns.despine()

#plt.savefig('shading_noshade.svg');
```

```python
plt.figure(figsize=(4, 4))
pd.Series(prices, index=dates).plot(marker='o', markersize=10)
plt.xticks([datetime(2019, 12, 1), datetime(2020, 1, 1), 
            datetime(2020, 2, 1),datetime(2020, 3, 1)], 
           ['Dec\n2019', 'Jan\n2020', 'Feb\n2020', 'March\n2020'], rotation=0, ha='center') 
plt.xlim(datetime(2019, 11, 10), datetime(2020, 3, 20))
plt.ylim(50, 150)
plt.axvspan(datetime(2019, 11, 10), datetime(2020, 1, 1), color='lightgray', alpha=0.5)

plt.text(datetime(2019, 12, 5), 140, '2019', fontsize=16, color='gray')
plt.text(datetime(2020, 1, 5), 140, '2020', fontsize=16, color='black')

sns.despine()

#plt.savefig('shading_withshade.svg');
```

# 5. Annotations

```python
earth_days_per_year = [
    ('Mercury', 88),
    ('Venus', 225),
    ('Earth', 365),
    ('Mars', 687),
    ('Jupiter', 4333),
    ('Saturn', 10759),
    ('Uranus', 30689),
    ('Neptune', 60182)
]
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(range(len(earth_days_per_year)), 
             [i[1] for i in earth_days_per_year], 
             s=150, facecolor='violet', 
             edgecolor='black'
            )
plt.ylabel('Days\nper Year', ha='left', position=(-.2, .8), rotation=0, labelpad=40)
plt.xticks([])
plt.yticks([0, 20000, 40000, 60000],  ['0', '20K', '40K', '60K'])

sns.despine()
plt.tight_layout()

#plt.savefig('annotations_noannot.svg');
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(range(len(earth_days_per_year)), 
             [i[1] for i in earth_days_per_year], 
             s=150, facecolor='violet', 
             edgecolor='black'
            )
plt.ylabel('Days\nper Year', ha='left', position=(-.2, .8), rotation=0, labelpad=40)
plt.xticks([])
plt.yticks([0, 20000, 40000, 60000],  ['0', '20K', '40K', '60K'])
plt.text(5.3, 55000, 'Neptune', fontsize=13, fontstyle='italic')
plt.text(4.5, 25000, 'Uranus', fontsize=13, fontstyle='italic')
plt.text(1.4, 4000, 'Earth', fontsize=13, fontstyle='italic')

sns.despine()
plt.tight_layout()

#plt.savefig('annotations_annot.svg');
```

# 6. LaTeX

```python
xx = np.linspace(-3, 3)
yy = xx**2 + 3
```

```python
plt.figure(figsize=(4, 4))
plt.plot(xx, yy, zorder=3, lw=3)
plt.ylim(0, None)
plt.axvline(0, ls='--', color='lightgray')
plt.title('\n')

sns.despine()
plt.tight_layout()

#plt.savefig('latex_notitle.svg');
```

```python
plt.figure(figsize=(4, 4))
plt.plot(xx, yy, zorder=3, lw=3)
plt.ylim(0, None)
plt.axvline(0, ls='--', color='lightgray')
plt.title(r'$\bf{y=x^2 + \beta}$', fontsize=24,  color='tab:red')

sns.despine()
plt.tight_layout()

#plt.savefig('latex_title.svg');
```
